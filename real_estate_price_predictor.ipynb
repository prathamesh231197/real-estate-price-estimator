```python
df1.drop(["PRICE_PER_UNIT_AREA"],axis=1,inplace=True)
```

```python
df1["AREA"] = df1["AREA"].str.replace("sq.ft.","")
```

```python
def convert_ranges_avg(x):
    tokens=x.split("-")
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None    
                     
```

```python
df1["AREA"] = df1["AREA"].apply(convert_ranges_avg)
```
